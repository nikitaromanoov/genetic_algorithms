```python
import random
import math
import pandas as pd
import numpy.random as npr
```

```python
def factory(N):
  paths = [i for i in range(N)]
  random.shuffle(paths) 
  return paths 
```

```python
def crossover(p1, p2):
    children = []
    z = [i for i in range(len(p1))]
    for i in range(len(p1)):
      if p1[i] == p2[i]:
        children +=[p1[i]]
        z.remove(p1[i])
      else:
        children += [-1]
    random.shuffle(z) 
    for i in range(len(p1)):
      if children[i] == -1:
        children[i] = z[0]
        del z[0]

    return children
```

```python
def fintess_function(comb):
  count_error = 0
  for i in range(1,len(comb)-1):
    for j in range(i+1,len(comb)):
      if (j-i) == math.fabs(comb[j] - comb[i]):
        count_error += 1

  return count_error
```

```python
def selection(population, populations_size):
  df = pd.DataFrame({"individ":population, "score": [fintess_function(i) for i in population]})
  df = df.sort_values("score")
  #df = df[~df.astype(str).duplicated()]
  return list(df["individ"][:populations_size]), min(list(df["score"][:populations_size]))
```

```python
def mutation(individs, COUNT):

  # swap mutation
  for j in range(COUNT):
      i = individs[j]
      rand = random.randrange(1, len(individs))
      random_1 = random.randrange(len(i))
      random_2 =  random.randrange(len(i))
      i[random_1], i[random_2] = i[random_2],i[ random_1]

  return individs
     
```

```python
def selectOne(population):
    max = sum(1/population.score)
    selection_probs = [(1/c)/max for c in population.score]
    return df["individ"][npr.choice(len(population), p=selection_probs)]
```

```python
GENERATIONS = 100000
POPULATIONS_SIZE = 300
SIZE = 256
PRECENT_CROSSOVER = 0.9
best_ever = 10000000
ELETISM = 30
COUNT_MUTATION = 50

population = []
for i in range(GENERATIONS):
  #factor
  if population == []:
    for j in range(POPULATIONS_SIZE):
      population += [factory(SIZE)]

  # elitism
  new_population = []
  df = pd.DataFrame({"individ":population, "score": [fintess_function(i) for i in population]})
  df = df.sort_values("score")
  base = list(df["individ"][:ELETISM])

  # crossover
  for j in range(round(PRECENT_CROSSOVER*round(POPULATIONS_SIZE/2))):
    new_population += [crossover(selectOne(df[~df.astype(str).duplicated()].reset_index()),selectOne(df))]
  population += new_population

  #mutation
  random.shuffle(population)
  population = mutation(population, COUNT_MUTATION)
  #selection
  population, best_score = selection(population, POPULATIONS_SIZE- ELETISM)
  population = base + population

  if best_score < best_ever:
    best_ever = best_score

  print(f"Generation #{i}, best_score_ever = {best_ever}, best_current_score = {best_score}")
  if best_score == 0:
    break
```
